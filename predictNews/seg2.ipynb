```python
#segmentation v2
import codecs
import csv
import pandas as pd
import jieba
import jieba.analyse

#data path
path = "C:\\Users\\user\\Desktop\\newscrawler\\outputcsv\\csvdata\\data2.csv"

#load data to dataframe
df = pd.read_csv(path, names = ["tag", "content"], usecols =[0,3], encoding='utf-8', dtype=str)

#pd.set_option('max_colwidth',100) 
print(df)
```

```python
cutdf = df.copy()

stopwords = [' ']
for word in open('C:\\Users\\user\\Desktop\\newscrawler\\keywordextraction\\stopwords.txt', 'r', encoding='utf-8'):
    stopwords.append(word.strip())

for index, row in cutdf.iterrows():
    row[1] = jieba.cut(row[1], cut_all=False)
    row[1] = " ".join(set(row[1])-set(stopwords))

pd.set_option("display.max_rows", None,'display.max_columns', None,'display.max_colwidth', None)
print(cutdf)
```
