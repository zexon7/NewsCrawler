```python
#segmentation
import codecs
import pandas as pd
import jieba
import jieba.analyse
from ckiptagger import WS, POS, NER
import tensorflow

#data path
path = "C:\\Users\\user\\Desktop\\newscrawler\\outputcsv\\csvdata\\data.csv"


df = pd.read_csv(path, names = ["tag", "title"], encoding='utf-8', dtype=str)
#pd.set_option('max_colwidth',100) 
print(df)
```

```python
#jieba version
cutdf = df.copy()
#keyword extraction
#df.title = [jieba.analyse.extract_tags(t, topK=10, withWeight=False, allowPOS=('ns', 'n', 'vn', 'v')) for t in df.title]

#stopwords
stopwords = [line.strip() for line in codecs.open('C:\\Users\\user\\Desktop\\newscrawler\\keywordextraction\\stopwords.txt', 'r', 'utf-8-sig').readlines()]

pd.set_option('max_colwidth',100)

for t in cutdf.title:
    t = jieba.cut(t, cut_all=False)
    a = " ".join(set(t)-set(stopwords))
    print(a)
```

```python
#ckiptagger version
cutdf = df.copy()
ws = WS(".\\CkipTagger_model")
stopwords = [line.strip() for line in codecs.open('C:\\Users\\user\\Desktop\\newscrawler\\keywordextraction\\stopwords.txt', 'r', 'utf-8-sig').readlines()]
c = ws(cutdf.title, sentence_segmentation=True)
#print(c)

# i for index,c for cutted words
for i in c:
    #print("Before remove punctuation marks and stopwords: ", i) 
    cc = []
    #for word in first index value
    for w in i:
        if w not in stopwords:
            cc.append(w)
    #print("After remove punctuation marks and stopwords: ", cc)
    print(cc)
```
